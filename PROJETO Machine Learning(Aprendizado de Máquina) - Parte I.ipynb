```python
import pandas as pd
arquivo = pd.read_csv('D:\\Temp\\CursoPythonDados\\wine_dataset.csv')
```

```python
arquivo ##ver dados
```

```python
##trocando os valores de vinhos para zero e um 
arquivo['style'] = arquivo['style'].replace('red',0) 
arquivo['style'] = arquivo['style'].replace('white',1).infer_objects(copy=False)
```

```python
##separando as variaveis a variavel 
## x vai trabalhar com todas as colunas menos a de vinhos=style
## y vai trabalhar apenas com a coluna de vinhos que no caso é a coluna style
y = arquivo ['style']
x = arquivo.drop('style', axis = 1)
```

```python
## verificando a mudança que fizemos no [comando 13] axis=0: Refere-se ao eixo das linhas. axis=1: Refere-se ao eixo das colunas. 
print (y)
```

```python
## aqui so mostra os outros dados da tabela separadaos 
print (x)
```

```python
## crinado trieno e teste para a maquina com a liby skalearn(treinando x e y) - vai treina  test_size=0.3 = 30% fo que temos e vai verificar 
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split (x,y, test_size=0.3)
```

```python
#o x se manteve com o restante das informações 
x_teste.shape
```

```python
## verificar acima que tiramos uma coluna
y_teste.shape
```

```python
## arvore de classificação
from sklearn.ensemble import ExtraTreesClassifier
#criando o modelo
modelo = ExtraTreesClassifier()
#chamando modelo e aplica o algoritimo
modelo.fit(x_treino, y_treino)
#imprimindo resultado
resultado = modelo.score(x_treino, y_treino)
print("probabilidade: ", resultado)
```
